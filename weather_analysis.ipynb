## Basic plotting

### Subtask:
Create basic plots to visualize the trends of temperature and rainfall over time.


**Reasoning**:
Create two plots to visualize the trends of maximum and minimum temperatures and rainfall over time using the 'Date', 'Tmax', 'Tmin', and 'Rain' columns of the dataframe.



```python
# Create a figure and axes object for the temperature plots
fig1, ax1 = plt.subplots(figsize=(12, 6))

# Plot the 'Tmax' column against the 'Date' column
ax1.plot(london_df['Date'], london_df['Tmax'], label='Max Temperature')

# Plot the 'Tmin' column against the 'Date' column
ax1.plot(london_df['Date'], london_df['Tmin'], label='Min Temperature')

# Set the title and labels for the temperature plot
ax1.set_title('Maximum and Minimum Temperatures Over Time')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)')

# Add a legend to the temperature plot
ax1.legend()

# Display the temperature plot
plt.show()

# Create a new figure and axes object for the rainfall plot
fig2, ax2 = plt.subplots(figsize=(12, 6))

# Plot the 'Rain' column against the 'Date' column
ax2.plot(london_df['Date'], london_df['Rain'], label='Rainfall', color='green')

# Set the title and labels for the rainfall plot
ax2.set_title('Rainfall Over Time')
ax2.set_xlabel('Date')
ax2.set_ylabel('Rainfall (mm)')

# Add a legend to the rainfall plot
ax2.legend()

# Display the rainfall plot
plt.show()
```

## Advanced plotting

### Subtask:
Create more informative visualizations, such as combining temperature and rainfall data or looking at distributions.


**Reasoning**:
Create a figure and axes with a secondary y-axis to plot temperature and rainfall on the same chart against the date.



```python
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot temperature data on the primary y-axis
ax1.plot(london_df['Date'], london_df['Tmax'], color='tab:red', label='Max Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (Â°C)', color='tab:red')
ax1.tick_params(axis='y', labelcolor='tab:red')

# Create a secondary y-axis for rainfall data
ax2 = ax1.twinx()
ax2.plot(london_df['Date'], london_df['Rain'], color='tab:blue', label='Rainfall')
ax2.set_ylabel('Rainfall (mm)', color='tab:blue')
ax2.tick_params(axis='y', labelcolor='tab:blue')

# Add title and legends
plt.title('Maximum Temperature and Rainfall Over Time')
fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))

plt.grid(True)
plt.show()
```
